Calculator in Python :)



```python
# asks the user to enter the first number
num1 = input("Enter the first number: ")

# asks the user to enter the second number
num2 = input("Enter the second number: ")

# asks the user to enter the operation to perform
print("Which operation do you want to perform?")
print("1. Addition")
print("2. Subtraction")
print("3. Multiplication")
print("4. Division")

# reads the operation chosen by the user
choice = input("Choose 1, 2, 3 or 4: ")

# performs the chosen operation
if choice == '1':
  result = float(num1) + float(num2)
  print(num1 + "+" + num2 + "=" + str(result))

elif choice == '2':
  result = float(num1) - float(num2)
  print(num1 + "-" + num2 + "=" + str(result))

elif choice == '3':
  result = float(num1) * float(num2)
  print(num1 + "*" + num2 + "=" + str(result))

elif choice == '4':
  result = float(num1) / float(num2)
  print(num1 + "/" + num2 + "=" + str(result))

else:
  print("Invalid choice")
```

Data mining on the students table

```python
# To create a students table you can use a data science library like Pandas and the Matplot lib library to generate the plots
import pandas as pd
import matplotlib.pyplot as plt
from operator import itemgetter

# create the students data table
students = {
    'Full_Name': ['Mario Rossi', 'Anna Bianchi', 'Luca Verdi','Sofia Ferrari'],
    'Subject': ['Mathematics', 'Italian', 'History', 'Computer Science'],
    'Date': ['01/01/2022', '01/02/2022', '01/03/2022', '01/04/2022'],
    'Grade': [7, 5, 6, 9]
}

# We create a DataFrame (two-dimensional data structure) with the students' data
df = pd.DataFrame(students)
# print the content of the DataFrame
print(df)
```

```python
# Once you've created the table, you can create plots to visualize the data in a more intuitive way. 
plt.title("Students' Grades") # Add the title to the plot
plt.xlabel("Students") # Add the label for the 'x' axis
plt.ylabel("Grades") # Add the label for the 'y' axis
df = df.sort_values(by='Grade', ascending=False) # Sort the DataFrame from the highest to the lowest grade
plt.bar(df['Full_Name'], df['Grade']) # Using Matplotlib you can create a bar plot to display the students' grades
plt.show() # Show the plot
```

```python
# Converts the values in the Grade column into a list
grades = df['Grade'].tolist()

# Create a dictionary that contains the number of students for each grade category (high, medium, low)
grade_categories = {"high": 0, "medium": 0, "low": 0}
for grade in grades:
    if grade >= 8:
        grade_categories["high"] += 1
    elif grade >= 6:
        grade_categories["medium"] += 1
    else:
        grade_categories["low"] += 1

# Extract the keys and values from the dictionary and assigns them to the labels and sizes variables
labels, sizes = zip(*grade_categories.items())

# Create the pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%')

# Add title
plt.title("Percentage of students per grade category")

# Show the chart
plt.show()
```

The hangman game

```python
# Import the random library to choose a word at random from a list of words
import random

# Define the list of words for the game
words = ["house", "tree", "cat", "phone", "computer", "monkey"]

# Choose a random word from the list of words
secret_word = random.choice(words)

# Create a variable to keep track of the letters correctly guessed by the user
guessed_letters = []

# Create a counter to keep track of the number of attempts left for the user
remaining_attempts = 6

# Create a while loop to continue until the user guesses the word or runs out of attempts
while True:
    # Create a variable to keep track of the current state of the game, showing the user the secret word with the missing letters
    state = ""
    for letter in secret_word:
        if letter in guessed_letters:
            state += letter
        else:
            state += "*"

    # Print the state of the game and the number of remaining attempts
    print(state)
    print(f"Remaining attempts: {remaining_attempts}")

    # Ask the user to enter a letter
    letter = input("Enter a letter: ")

    # Check if the entered letter is present in the secret word
    if letter in secret_word:
        # If the letter is present, add it to the list of guessed letters
        guessed_letters.append(letter)

        # Check if the user has guessed the secret word
        if all(letter in guessed_letters for letter in secret_word):
            # If the user has guessed the secret word, print a congratulations message and break the while loop
            print("Congratulations, you guessed the secret word!")
            break
    else:
        # If the letter is not present, decrement the counter of remaining attempts
        remaining_attempts -= 1

        # Check if the user has run out of attempts
        if remaining_attempts == 0:
            # If the user has run out of attempts, print a defeat message and end the game
            print("Sorry, you lost!")
            break
```

```python
# Import Matplotlib library
import matplotlib.pyplot as plt

# Data to be represented in the chart
grades = [7, 10, 6, 8, 7, 9, 8, 9, 6, 7]

# Create a dictionary that contains the number of students for each grade
grades_counted = {}
for grade in grades:
    if grade in grades_counted:
        grades_counted[grade] += 1
    else:
        grades_counted[grade] = 1

# Extract keys and values from the dictionary and assign them to the variables x and y
x, y = zip(*grades_counted.items())

# Create the bar chart
plt.bar(x, y)

# Add title and labels for the axes
plt.title("Number of students per grade")
plt.xlabel("Grade")
plt.ylabel("Number of students")

# Show the chart
plt.show()
```
